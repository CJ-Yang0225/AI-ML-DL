假設我們有兩組數字，分別為X、Y，觀察兩組數字之間的關係可以發現一個"規則"。
```
y = (2 * x) - 1
```
但並非所有數據都能輕易找出規律，這時就可以透過數據，訓練一個神經網絡來找出它們之間的關係，達成我們的需求。

````
import tensorflow as tf
import numpy as np
from tensorflow import keras
````

TensorFlow中有一個稱為keras的API，能使定義神經網絡變得容易許多。

````
# 建立只有一層、一個神經元、輸入和輸出為一個值的簡單神經網絡
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````

接著選定損失函數(Loss Function)及優化器(Optimizer)。

````
# 均方誤差(MSE)、隨機梯度下降(SGD)
model.compile(loss='mean_squared_error', optimizer='SGD')
````

給予分別有6個數字的資料

````
Xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
Ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

````
# 訓練模型，使模型找出適合Y的X值，epochs為循環次數
model.fit(Xs, Ys, epochs=500)
````

神經網絡一開始不知道X和Y之間的關係，它會先隨機猜測一個函數，像是：
```
y = (10 * x) - 10
```
然後帶入資料，返回結果，損失函數對此評估好壞，傳送給優化器，再進行下一個猜測，慢慢提升準確度。

模型訓練完成，輸入一個未知的資料讓模型預測。

````
print(model.predict([10.0]))
````

我們一般認為(2*10)-1=19，但結果不是，主要原因：

訓練的資料過少，只有6筆，神經網絡計算答案是透過機率來處理，算出X和Y之間的關係高機率是Y=2X-1，因此結果非常接近19，但不一定是19。
